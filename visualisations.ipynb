```python
import pandas as pd
import plotly.express as ex
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('runs/RecognitionNet-iresnet50-MagFaceLoss-ciWLXb/csvs/metrics.csv')
df_train = df.dropna(subset=['train_acc', 'train_loss', 'step'])
df_val = df.dropna(subset=['val_acc', 'val_loss', 'step', 'epoch'])
```

```python
fig = make_subplots(2,3)
fig.add_traces(
    [
    go.Scatter(df_train,x='step', y='train_acc'),
    go.Scatter(df_val,x='step', y='val_acc'),
    ]
    rows=[1, 1],
    cols=[1, 2]
)
fig1 = ex.line(df_train, x='step', y='train_acc')
fig2 = ex.line(df_train, x='step', y='train_loss')
fig3 = ex.line(df_val, x='step', y='val_acc')
fig4 = ex.line(df_val, x='step', y='val_loss')
fig5 = ex.line(df_val, x='step', y='epoch')
fig1.show()
fig2.show()
fig3.show()
fig4.show()
fig5.show()
```
